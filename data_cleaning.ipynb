```python
import pandas as pd
```

```python
file_id = '1tZnTtKT1sRm1uFK0Gs7FgKqreY9Mruw5'
url = f'https://drive.google.com/uc?id={file_id}&export=download'
df = pd.read_excel(url)
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
```

`Removing` symbols from `Last_Name`

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

`Cleaning/Standardizing` phone number

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r'\D','', regex=True) # removing symbols from phone numbers
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) # truning float to string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # setting dash between numbers
df["Phone_Number"] = df["Phone_Number"].str.replace('--','')
```

`Splitting` Columns

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

`Standardizing` Column Values using `replace`

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

```python
df = df.replace('N/a', '')
df = df.replace('nan', '')
df = df.fillna("")
df
```

`Filtering` down rows of data

```python
for x in df.index:
  if df.loc[x, "Do_Not_Contact"] == 'Y':
    df.drop(x, inplace=True)
df
```

```python
#df.dropna(subset=['Phone_Number'], inplace=True)

for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace=True)
df
```

```python
df = df.reset_index(drop=True)
df
```
